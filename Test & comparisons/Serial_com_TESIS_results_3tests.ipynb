```python
#ARCHIVO PARA LA COMUNICACIÓN SERIAL ENTRE EL CIRCUITO Y EL CÓDIGO

#%%
from joblib import load
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
import serial  # PARA LA COMUNICACIÓN SERIAL

#Carga de datos de entrenamiento para calcular la media y el estándar
train_x = np.load('c:\\Users\\Indatech\\Downloads\\EMG\\train_x.npy')
```

```python
#%%

# PARA NORMALIZAR
mean = train_x.mean(axis=0)
std = train_x.std(axis=0)
```

```python
#%%

#Cargando modelo KNN
KNN_model = load('c:\\Users\\Indatech\\Downloads\\EMG\\knn_model.joblib')

#ser = serial.Serial('COM4', 9600) 

arduino = serial.Serial(port='COM3', baudrate=9600, timeout=.1) 
```

```python
# PRUEBA #1 

# PREDICTIONS:
    #0 - Unknown,
    #1 - hand at rest,
    #2 - hand clenched in a fist,
    #3 - wrist flexion,
    #4 – wrist extension

positions = ['Mano en reposo', 'Mano en puño', 'Flexión de muñeca', 'Extensión de muñeca']

i = 0
while True:
    # Lectura de datos EMG sin procesar desde el puerto serial
    raw_data = arduino.readline()
    
    print('Data no procesada: ', raw_data)

    # Preprocesando la data
    arduino_to_str = raw_data.decode('utf-8')
    str_to_list = arduino_to_str.split(",") #Separa la cadena en una lista de valores basada en las comas
    voltage = str_to_list[1]  #Extrayendo el valor que quiero (los otros son el rango 1-5 V)
    voltage = float(voltage)  #Convertir de str a float
    voltage = voltage*0.01
    print('Voltaje: ', voltage)

    # Poniendo en el formato que necesitamos para hacer la predicción
    voltage_array = np.repeat(voltage, 4) # 4 canales
    print(voltage_array)
    voltage_array = np.reshape(voltage_array, (1,4))
    voltage_array -= mean
    voltage_array /= std

    print(voltage_array)

    if voltage == 0.0:
        print('Predicción: No hay movimiento')  #AJUSTE MANUAL
    # Realizando la prediccion
    prediction = KNN_model.predict(voltage_array)  # Predice usando el modelo cargado

    # Salida
    if prediction[0] != 0 :
        print('Predicción: ', positions[prediction[0]-1])  
    else:
        print('Predicción: ', positions[prediction[0]])

    i = i+1
    #if i > 45:
     #   x = input('Continuar con la adquisición de señales?  (S/N) ')
      #  if x.lower() == 'n':
       #     break
```
